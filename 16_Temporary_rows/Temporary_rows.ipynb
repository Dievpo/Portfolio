#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

### Тренды и сезонность 

Чтобы лучше понимать временной ряд, разберём тренды и сезонность

```python
data = data.drop(['mean','std', 'rolling_mean'], axis=1)
```

```python
# функция seasonal_decompose() (англ. «разбить на части»). Она раскладывает временной ряд на три составляющие: тренд, сезонность и остаток (англ. residuals).
decomposed = seasonal_decompose(data) 

plt.figure(figsize=(12, 10)) 
plt.subplot(311) 

# Чтобы график корректно отобразился, указываем его оси ax, равными plt.gca() 

# Trend                                                                       
decomposed.trend.plot(ax=plt.gca()) 
plt.title('Trend') 
plt.subplot(312) 

# Seasonality 
decomposed.seasonal.plot(ax=plt.gca()) 
plt.title('Seasonality') 
plt.subplot(313) 

# Residuals
decomposed.resid.plot(ax=plt.gca()) 
plt.title('Residuals') 
plt.tight_layout() 
```

На данном графике заметен глобальный тренд увеличения числа заказов такси на протяжении всего изучаемого периода.

Видимо, в этот момент прилетает много самолетов и отсустствует дополнительный наземный транспорт, которым можно добраться из аэропорта.
Минимальное количество заказов в районе 6-7 часов утра.
Так же заметен второй по величине пик заказов в районе 16-17 дня.

## Обучение

### Подготовка признаков

На вход функция принимает датасет, количество "шагов" для отстающих значений и скользящее среднее без учета текущего значения количества заказов.  
По результатам исследования мы увидели, что у нас есть дневная и недельная сезонность. Соответственно, из лагов нам пригодится лаг предыдущего часа, того же и предыдущего часа предыдущего дня, недельной давности. 

```python
def new_features(data, max_lag, rolling_mean_size):
    
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

Значимость лагов можно посмотреть по автокорреляции. Функция показывает, как признак коррелирует с предыдущими. Область синей заливки обозначает порог статистической значимости (если значение внутри, корреляция уже не заслуживает доверия. Статзначимость вычисляется по тесту Стьюдента).

```python
from statsmodels.graphics import tsaplots

taxi = pd.read_csv("/datasets/taxi.csv", index_col=[0], parse_dates=[0]).resample('1H').sum()
ts = taxi['num_orders']

fig = tsaplots.plot_acf(ts,lags= 190)
fig.set_size_inches(18.5, 6)
plt.ylim(-.1, 1.1)
plt.show() 
```

Поэтому возьмем первые 25 лагов

```python
new_features(data, 25, 30)
```

### Создание подвыборок

Разобьем датасет на обучающую и тестовую выборку.

```python
data.dropna(inplace=True)
```

```python
features = data.drop(['num_orders'], axis=1) # выделение признаков
target = data['num_orders']
```

```python
features_train, features_test, target_train, target_test = train_test_split(
    features, target, shuffle=False, test_size=0.1)
```

```python
features_train.shape[0], features_test.shape[0], target_train.shape[0], target_test.shape[0]
```

### Обучение моделей

Обучим различные модели и добавим модели и рассчитанный RMSE в массив

```python
train_models = []
```

#### Линейная регрессия

```python
%%time
lr = LinearRegression()
lr.fit(features_train, target_train)
cv = TimeSeriesSplit(n_splits=3).split(features_train)
score_lr = cross_val_score(lr, features_train, target_train, scoring='neg_mean_squared_error', cv=cv) 
rmse = round(np.mean(-score_lr)**0.5, 2)
rmse
```

```python
train_models.append([lr, rmse])
```

#### DecisionTreeRegressor

```python
%%time
dt = DecisionTreeRegressor(random_state=12345)

parameters = {'max_depth':range(1,11,2)}

tscv = TimeSeriesSplit(n_splits=2)

gs_dt = GridSearchCV(estimator=dt, cv=tscv, param_grid=parameters, scoring = 'neg_mean_squared_error')

gs_dt.fit(features_train, target_train)
print(gs_dt.best_params_)

rmse = round(np.mean(-gs_dt.best_score_)**0.5, 2)
print(rmse)
```

```python
train_models.append([gs_dt, rmse])
```

#### RandomForestRegressor

```python
%%time
rf = RandomForestRegressor(random_state=12345)

parameters = {'n_estimators':range(50,100,10), 'max_depth':range(1,11,2)}

tscv = TimeSeriesSplit(n_splits=2)

gs_rf = GridSearchCV(estimator=rf, cv=tscv, param_grid=parameters, scoring = 'neg_mean_squared_error')
gs_rf.fit(features_train, target_train)
print(gs_rf.best_params_)

rmse = round(np.mean(-gs_dt.best_score_)**0.5, 2)
print(rmse)
```

```python
train_models.append([gs_rf, rmse])
```

#### CatBoost

```python
%%time
cat = CatBoostRegressor(verbose=False)

cat.fit(features_train, target_train, verbose=False)
cv = TimeSeriesSplit(n_splits=3).split(features_train)
score_cat = cross_val_score(cat, features_train, target_train, scoring='neg_mean_squared_error', cv=cv) 

rmse = round(np.mean(-score_cat)**0.5, 2)
rmse
```

```python
train_models.append([cat, rmse])
```

#### LightGBM

```python
%%time
lg = lgb.LGBMRegressor()

lg.fit(features_train, target_train, verbose=False)
cv = TimeSeriesSplit(n_splits=3).split(features_train)
score_lg = cross_val_score(lg, features_train, target_train, scoring='neg_mean_squared_error', cv=cv) 

rmse = round(np.mean(-score_lg)**0.5, 2)
rmse
```

```python
train_models.append([lg, rmse])
```

### Результаты

```python
result = pd.DataFrame(train_models, columns=['model', 'rmse_train'])
```

```python
pd.options.display.max_colwidth = 0
```

```python
display(result.sort_values(by='rmse_train'))
```

## Тестирование

Протестируем обученные модели на тестовых данных и добавим их RMSE в датафрейм с результатами, а также построим графики исходного и предсказанных рядов для каждой модели

```python
test_models = []
```

```python
sns.set(rc={'figure.figsize':(15,4)})
def display_result(target, pred, rmse):
    result = target.to_frame().reset_index()
    result['prediction'] = pd.Series(pred)
    result.set_index('datetime', inplace=True)
    result.plot()
    plt.title('RMSE='+str(rmse))
    plt.show()
```

```python
for i in range(len(train_models)):
    predictions = train_models[i][0].predict(features_test)
    rmse = round(mean_squared_error(target_test, predictions, squared=False), 2)
    display_result(target_test, predictions, rmse)
    test_models.append(rmse)
```

CatBoostRegressor показывает хорошие результаты - график предсказаний более похож на исходные значения, чем у остальных моделей, но не очень хорошо предсказывает силу скачков, но сам факт повышенного спроса мы ловим почти всегда верно. RMSE = 40,78. Это означает, что модель в среднем ошибаетс на 40-41 заказ такси в час.

```python
result['rmse_test'] = pd.Series(test_models)
```

```python
result.sort_values(by='rmse_test')
```

**Проверим модели на адекватность**

```python
pred_previous = target_test.shift()
pred_previous.iloc[0] = target_train.iloc[-1]

rmse = round(mean_squared_error(target_test, pred_previous, squared=False), 2)
rmse
```

Все значения ниже чем 58.93 (тк 58.93 - это ориентир). Проверка на адекватность пройдена

**Вывод:**  
Все модели кроме DecisionTreeRegressor демонстрируют RMSE ниже 48.  Лучше всего показывает себя CatBoostRegressor со значением rmse=40.78
